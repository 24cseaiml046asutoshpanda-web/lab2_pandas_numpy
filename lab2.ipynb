```python
# it is using the library like numpy and pandas for finding the mean, mode, median for a given group of data in the last

import numpy as np
import pandas as pd

l1 = list(eval(input("Enter the data set : ")))
series = pd.Series(l1)
# print(series)

mean = np.mean(series)
print("Mean : ",mean)
median = np.median(series)
print("Median : ",median)

mode = series.mode()
print("Mode : ",mode)
```

```python
# it is for finding the mean, mode, and median for ungroup data with out using the inbuilt library

n = len(l1)
# Mean
mean1 = sum(l1) / n
sorted_data = sorted(l1)

# Median Logic
if n % 2 == 0:
    # Average of the two middle elements
    m1 = sorted_data[n // 2 - 1]
    m2 = sorted_data[n // 2]
    median1 = (m1 + m2) / 2
else:
    # The middle element
    median1 = sorted_data[n // 2]

# Mode Logic
frequency = {}
for i in l1:
    frequency[i] = frequency.get(i, 0) + 1

max_freq = max(frequency.values())
# Find all values that match the max frequency (handles multimodal data)
mode1 = [k for k, v in frequency.items() if v == max_freq]

print("Mean   :", mean1)
print("Median :", median1)
print("Mode   :", mode1)
```

```python
try:
    num_classes = int(input("Enter the number of class interval : "))
    classes_data = []
    
    print("Enter the details (Lower limit, Upper limit, frequency) : ")
    for i in range(num_classes):
        datas = input(f"Class {i+1} - Enter low, high, freq: ")
        parts = datas.replace(",", " ").split()
        low, high, freq = map(int, parts)
        classes_data.append({'low': low, 'high': high, 'freq': freq})

    fi_x = 0
    fi = 0   
    cumulative_freqs = []
    for ci in classes_data:
        midpoint = (ci['low'] + ci['high']) / 2
        fi_x += midpoint * ci['freq']
        fi += ci['freq']
        cumulative_freqs.append(fi) # This list is your 'cf' column
        
    # --- MEAN ---
    if fi != 0:
        mean2 = fi_x / fi
        print(f"\nGrouped Mean: {mean2}")

        # --- MEDIAN ---
        n_by_2 = fi / 2
        median_class_index = 0
        
        # 1. Find the Median Class
        for i in range(len(cumulative_freqs)):
            if cumulative_freqs[i] >= n_by_2:
                median_class_index = i
                break
        
        # 2. Extract values for the formula
        l = classes_data[median_class_index]['low']
        f = classes_data[median_class_index]['freq']
        h = classes_data[median_class_index]['high'] - classes_data[median_class_index]['low']
        
        # cf is the cumulative frequency of the PREVIOUS class
        if median_class_index == 0:
            prev_cf = 0
        else:
            prev_cf = cumulative_freqs[median_class_index - 1]
            
        # 3. Apply Median Formula: l + ((n/2 - prev_cf) / f) * h
        median2 = l + (((n_by_2 - prev_cf) / f) * h)
        print(f"Grouped Median: {median2}")

    else:
        print("Total frequency is zero; cannot calculate results.")

except (ValueError, IndexError, ZeroDivisionError) as e:
    print(f"Error: {e}")
```
