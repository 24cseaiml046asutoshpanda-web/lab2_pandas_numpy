```python
# it is using the library like numpy and pandas for finding the mean, mode, median for a given group of data in the last

import numpy as np
import pandas as pd

l1 = list(eval(input("Enter the data set : ")))
series = pd.Series(l1)
# print(series)

mean = np.mean(series)
print("Mean : ",mean)
median = np.median(series)
print("Median : ",median)

mode = series.mode()
print("Mode : ",mode)
```

```python
# it is for finding the mean, mode, and median for ungroup data with out using the inbuilt library

n = len(l1)
# Mean
mean1 = sum(l1) / n
sorted_data = sorted(l1)

# Median Logic
if n % 2 == 0:
    # Average of the two middle elements
    m1 = sorted_data[n // 2 - 1]
    m2 = sorted_data[n // 2]
    median1 = (m1 + m2) / 2
else:
    # The middle element
    median1 = sorted_data[n // 2]

# Mode Logic
frequency = {}
for i in l1:
    frequency[i] = frequency.get(i, 0) + 1

max_freq = max(frequency.values())
# Find all values that match the max frequency (handles multimodal data)
mode1 = [k for k, v in frequency.items() if v == max_freq]

print("Mean   :", mean1)
print("Median :", median1)
print("Mode   :", mode1)
```
